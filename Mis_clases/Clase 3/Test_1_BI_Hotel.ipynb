### 4.2 The IQR Method – Detecting Outliers

**Formula:**
- Q1 = 25th percentile, Q3 = 75th percentile
- IQR = Q3 − Q1
- Lower bound = Q1 − 1.5 × IQR
- Upper bound = Q3 + 1.5 × IQR

Any value outside these bounds is flagged as an outlier.
