## 📦 Section 4 – Outlier Detection & Treatment

### 4.1 Visualise with a Box Plot

Box plots are the fastest way to spot outliers.

```python
# YOUR CODE HERE: Generate a box-plot for room_rate variable.

sns.boxplot(data = df, y = df['room_rate'])
plt.show()
```

### 4.2 The IQR Method – Detecting Outliers

**Formula:**
- Q1 = 25th percentile, Q3 = 75th percentile
- IQR = Q3 − Q1
- Lower bound = Q1 − 1.5 × IQR
- Upper bound = Q3 + 1.5 × IQR

Any value outside these bounds is flagged as an outlier.

```python
# YOUR CODE HERE
# Compute IQR bounds AND print how many outliers were detected for room_rate variable.

Q1 = df['room_rate'].quantile(0.25)
Q2 = df['room_rate'].quantile(0.50)
Q3 = df['room_rate'].quantile(0.75)
IQR = Q3-Q1
LB = Q1-1.5*IQR
UB = Q3+1.5*IQR

df[(df['room_rate'] < LB) | (df['room_rate'] > UB)].reset_index()
```

### 4.3 Winsorization – Capping Outliers

Instead of removing outliers, **cap them** at the 5th and 95th percentile so the record stays in the dataset.


```python
# YOUR CODE HERE: Apply Winsorization for room_rate column
upper_bound = df['room_rate'].quantile(0.95)
lower_bound = df['room_rate'].quantile(0.05)

df['room_rate'] = df['room_rate'].clip(lower_bound, upper_bound)
df.describe()
```

### 4.4 Log Transformation

Log transformation compresses large values, making skewed distributions more symmetric.

```python
# YOUR CODE HERE: Apply it for room_rate variable. Create a new column called: room_rate_log, that contains the transformed value.

df['room_rate_log'] = np.log10(df['room_rate'])
df.head()
```
