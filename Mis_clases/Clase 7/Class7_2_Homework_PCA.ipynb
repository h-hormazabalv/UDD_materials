# Principal Component Analysis (PCA) — Case Study 2

## Employee Retention Risk at TechNova

---

## 1. Business Context

**TechNova** is a mid-size software company with roughly 500 employees. The HR analytics team collected behavioral and organizational data on **n = 150 employees** with three goals:

1. **Dimensionality reduction:** compress 10 correlated workforce variables into a smaller set of interpretable dimensions.
2. **Visualization:** project employees onto the principal component space to discover behavioral profiles and risk groups.
3. **Regression:** predict a Retention Risk Score (0–100, where 100 = very likely to leave) using the original variables versus the principal components, and compare model performance.

---

## 2. Variable Description

The dataset contains 10 predictor variables organized in four organizational domains, plus one response variable.

### Compensation & Growth

| Variable | Type | Range | Description |
|----------|------|-------|-------------|
| `monthly_salary` | Continuous | 2,000–12,000 USD | Gross monthly salary |
| `salary_increase_pct` | Continuous | 0–20 % | Most recent annual salary increase as a percentage |
| `promotions_5yr` | Discrete | 0–5 | Number of promotions received in the last five years |

### Workload & Stress

| Variable | Type | Range | Description |
|----------|------|-------|-------------|
| `weekly_hours` | Discrete | 30–70 | Average number of hours worked per week |
| `overdue_tasks` | Discrete | 0–20 | Number of currently overdue tasks or deliverables |
| `sick_days_year` | Discrete | 0–25 | Sick days taken in the past 12 months |

### Engagement & Culture

| Variable | Type | Range | Description |
|----------|------|-------|-------------|
| `peer_collaboration` | Ordinal | 1.0–10.0 | Peer-rated collaboration score from 360° review |
| `initiative_count` | Discrete | 0–15 | Number of voluntary initiatives or improvement proposals submitted in the past year |

### Tenure & Stability

| Variable | Type | Range | Description |
|----------|------|-------|-------------|
| `years_at_company` | Continuous | 0.5–20 | Total years employed at TechNova |
| `projects_completed` | Discrete | 1–40 | Cumulative number of projects delivered |

### Response Variable

| Variable | Type | Range | Description |
|----------|------|-------|-------------|
| `retention_risk` | Continuous | 0–100 | Composite risk index estimated from exit-interview models, internal surveys, and historical turnover patterns. A score of 100 indicates the employee is very likely to leave within 12 months. |
