```python
import pandas as pd
import numpy as np

np.random.seed(42)
n_samples = 1000

# 1. Metros Cuadrados (Relación logarítmica: el precio por m2 baja a mayor tamaño)
m2 = np.random.normal(70, 30, n_samples).clip(20, 250)

# 2. Antigüedad (Años)
antiguedad = np.random.uniform(0, 50, n_samples)

# 3. Distancia al Centro Financiero (km)
distancia_centro = np.random.uniform(0.5, 15, n_samples)

# 4. Calidad de Acabados (Categoría numérica 1-5)
calidad = np.random.choice([1, 2, 3, 4, 5], n_samples)

# 5. Piso del departamento (Importante para el efecto "escalón")
piso = np.random.randint(1, 20, n_samples)

# --- Generación del Target (Precio en miles de USD) ---
# Introducimos NO LINEALIDADES:
# - El precio sube con m2 pero se estabiliza (raíz cuadrada).
# - La distancia al centro penaliza fuertemente si es < 3km (efecto campana).
# - Hay un salto de precio ("step") si el acabado es > 3.

precio = (
    (15 * np.sqrt(m2)) +                   # No lineal: m2
    (30 * np.exp(-distancia_centro/5)) +   # No lineal: cercanía centro
    (-0.5 * antiguedad) +                  # Lineal: depreciación
    (10 * (calidad > 3)) +                 # Salto: Acabados premium
    (2 * np.sin(piso)) +                   # Fluctuación: Preferencia por ciertos pisos
    np.random.normal(50, 5, n_samples)     # Base + Ruido
)

df = pd.DataFrame({
    'Metros_Cuadrados': m2.round(1),
    'Antiguedad': antiguedad.round(0),
    'Distancia_Centro_KM': distancia_centro.round(2),
    'Calidad_Acabados': calidad,
    'Piso': piso,
    'Precio_USD_k': precio.round(2)
})

print(df.head())
```
