# **Class 1 - Business Intelligence**

Now we get to the fun part — hands-on Python. We'll be using Google Colab, which is a free, browser-based Python environment. You don't need to install anything. If you can open a browser, you can run code today.

Here's what we'll cover in this practical session. First, we'll do a quick Python refresher: **variables, lists, dictionaries, loops, and functions** — the building blocks you'll use throughout the course.

**1. Variables**
What they are: Containers that store a value under a name.

**2. Lists**
What they are: Ordered collections of items — like a column in a spreadsheet.

**3. Dictionaries**
What they are: Key-value pairs — like a mini database row.

**4. Loops**
What they are: Repeat an action for every item in a collection.

**5. Functions**
What they are: Reusable blocks of code — write once, use many times.

**6. Create a DataFrame**
What it is: A table with rows and columns — like Excel, but in Python.
